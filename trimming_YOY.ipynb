```python
# Dependencies
import pandas as pd
import datetime as dt
```

```python
# Save path to data set in a variable
data_file = "Resources/YoY_Seated_Diner_Data.csv"
```

```python
# Use Pandas to read data
data_file_df = pd.read_csv(data_file)
data_file_df.head(25)
```

```python
# Get a list of all of our columns for easy reference
data_file_df.columns
```

```python
# Remove anything listed as country
states_df = data_file_df.loc[(
    data_file_df["Type"] == "state")]
states_df.head()
```

```python
#need only US states
states_df["Name"].unique
```

```python
us_states = ["Arizona","California","Colorado","Connecticut","Delaware","Florida","Georgia","Hawaii","Illinois","Indiana","Kansas","Kentucky","Louisiana","Maine","Maryland","Massachusetts","Michigan","Minnesota","Missouri","Nebraska","Nevada","New Hampshire","New Jersey","New Mexico","New York","North Carolina","Ohio","Oklahoma","Oregon","Pennsylvania","Rhode Island","South Carolina","Tennessee","Texas","Utah","Washington","Wisconsin"]
len(us_states)
```

```python
#The following states are not in this dataset:
# Alabama
# Alaksa
# Arkansas
# Idaho
# Iowa
# Mississippi
# Montana
# North Dakota
# South Dakota
# Vermont
# Virginia
# West Virginia
# Wyoming
```

```python
#if name is a state, write to new df

us_states_df=states_df[states_df["Name"].isin(us_states)]
us_states_df
    
```

```python
# unpivot dates
# us_states_df
date = list(us_states_df.columns[2:])
full_date = []

for day in date:
        day = day +'/20'
        full_date.append(day)

states = []
data = []
dates =[]


for index, rows in us_states_df.iterrows():
    if len(rows.tolist()[2:]) == len(date):
        data = data + rows.tolist()[2:]
        state = rows[1]
        states = states + [state]*len(date)
        dates = dates + full_date
        
        
# create new clean df
yoy_states = pd.DataFrame({
    'State':states,
    'Day': dates,
    'YOYPctChange': data
})
```

```python
# # Push the remade DataFrame to a new CSV file
yoy_states.to_csv("Resources/YOY_OpenTable states.csv",index=False)
```
